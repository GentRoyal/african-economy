```python
import pandas as pd
import numpy as np

def clean_economic_data(df):
    """
    Thoroughly clean the economic dataset based on identified data quality issues.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        The raw economic data
        
    Returns:
    --------
    pandas.DataFrame
        Cleaned dataset
    dict
        Cleaning report with statistics
    """
    df_clean = df.copy()
    initial_rows = len(df_clean)
    
    cleaning_report = {
        'initial_rows': initial_rows,
        'rows_removed': {},
        'standardizations': {}
    }
    
    # ==================== CLEAN COUNTRY ====================
    # Remove rows with missing Country
    null_countries = df_clean['Country'].isna().sum()
    df_clean = df_clean[df_clean['Country'].notna()]
    cleaning_report['rows_removed']['null_country'] = null_countries
    
    # Strip whitespace
    df_clean['Country'] = df_clean['Country'].str.strip()
    
    
    # ==================== CLEAN INDICATOR ====================
    # Remove rows with missing Indicator
    null_indicators = df_clean['Indicator'].isna().sum()
    df_clean = df_clean[df_clean['Indicator'].notna()]
    cleaning_report['rows_removed']['null_indicator'] = null_indicators
    
    # Strip whitespace from indicators
    df_clean['Indicator'] = df_clean['Indicator'].str.strip()
    
    # Standardize indicator names
    indicator_mapping = {
        'GDP per capita': 'GDP per Capita',
        'GDP per Capita ': 'GDP per Capita',
        'Inflation Rate ': 'Inflation Rate',
        'Food Inflation ': 'Food Inflation'
    }
    df_clean['Indicator'] = df_clean['Indicator'].replace(indicator_mapping)
    cleaning_report['standardizations']['indicators'] = indicator_mapping
    
    
    # ==================== CLEAN SOURCE ====================
    # Handle missing sources (keep rows but mark as Unknown)
    null_sources = df_clean['Source'].isna().sum()
    df_clean['Source'] = df_clean['Source'].fillna('Unknown')
    df_clean['Source'] = df_clean['Source'].str.strip()
    cleaning_report['standardizations']['null_sources_filled'] = null_sources
    
    
    # ==================== CLEAN UNIT ====================
    # Standardize Unit values
    unit_mapping = {
        'Millions': 'Million',
        'Billions': 'Billion',
        'points': 'Points',
        'Persons/%': 'Percent'
    }
    df_clean['Unit'] = df_clean['Unit'].replace(unit_mapping)
    df_clean['Unit'] = df_clean['Unit'].str.strip() if df_clean['Unit'].dtype == 'object' else df_clean['Unit']
    cleaning_report['standardizations']['units'] = unit_mapping
    
    
    # ==================== CLEAN CURRENCY ====================
    # Currency can be null for non-monetary indicators (like %, Points, Persons)
    # For monetary values, try to infer from Country
    
    # First, create a mapping of countries to their primary currencies
    country_currency_map = df_clean[df_clean['Currency'].notna()].groupby('Country')['Currency'].agg(
        lambda x: x.value_counts().index[0] if len(x) > 0 else None
    ).to_dict()
    
    def fill_currency(row):
        if pd.notna(row['Currency']):
            return row['Currency'].strip()
        
        # If Currency is missing, infer based on Unit type
        if row['Unit'] in ['%', 'Points', 'Percent']:
            return 'N/A'  # Not applicable for percentages/indices
        elif row['Unit'] == 'Persons':
            return 'N/A'  # Not applicable for population counts
        elif row['Unit'] == 'USD':
            return 'USD'  # Unit itself indicates USD
        elif row['Unit'] in ['Million', 'Billion']:
            # For monetary values, try to use country's primary currency
            if row['Country'] in country_currency_map:
                return country_currency_map[row['Country']]
            else:
                return 'Unknown'
        else:
            return 'Unknown'
    
    # Track currency fills before applying
    currency_fill_summary = df_clean[df_clean['Currency'].isna()].groupby('Unit')['Unit'].count().to_dict()
    
    # Apply currency filling
    df_clean['Currency'] = df_clean.apply(fill_currency, axis=1)
    
    # Count how many monetary values still have Unknown currency
    unknown_monetary = df_clean[
        (df_clean['Currency'] == 'Unknown') & 
        (df_clean['Unit'].isin(['Million', 'Billion']))
    ].shape[0]
    
    cleaning_report['standardizations']['currency_fills_by_unit'] = currency_fill_summary
    cleaning_report['standardizations']['monetary_values_inferred'] = (
        currency_fill_summary.get('Million', 0) + 
        currency_fill_summary.get('Billion', 0) + 
        currency_fill_summary.get('USD', 0) - 
        unknown_monetary
    )
    cleaning_report['warnings'] = {
        'unknown_monetary_currencies': unknown_monetary
    }
    
    
    # ==================== CLEAN FREQUENCY ====================
    # Remove rows with missing Frequency (important for time series)
    null_frequency = df_clean['Frequency'].isna().sum()
    df_clean = df_clean[df_clean['Frequency'].notna()]
    df_clean['Frequency'] = df_clean['Frequency'].str.strip()
    cleaning_report['rows_removed']['null_frequency'] = null_frequency
    
    
    # ==================== CLEAN COUNTRY CODE ====================
    # Remove rows with missing Country Code
    null_codes = df_clean['Country Code'].isna().sum()
    df_clean = df_clean[df_clean['Country Code'].notna()]
    df_clean['Country Code'] = df_clean['Country Code'].str.strip()
    cleaning_report['rows_removed']['null_country_code'] = null_codes
    
    
    # ==================== CLEAN TIME ====================
    # Remove rows with missing Time
    null_time = df_clean['Time'].isna().sum()
    df_clean = df_clean[df_clean['Time'].notna()]
    cleaning_report['rows_removed']['null_time'] = null_time
    
    # Convert to datetime
    df_clean['Time'] = pd.to_datetime(df_clean['Time'], errors='coerce')
    
    # Remove any rows where datetime conversion failed
    failed_time = df_clean['Time'].isna().sum()
    df_clean = df_clean[df_clean['Time'].notna()]
    cleaning_report['rows_removed']['invalid_time'] = failed_time
    
    
    # ==================== CLEAN AMOUNT ====================
    # Convert Amount to numeric
    df_clean['Amount'] = pd.to_numeric(df_clean['Amount'], errors='coerce')
    
    # Remove rows where Amount is null (can't have economic data without values)
    null_amount = df_clean['Amount'].isna().sum()
    df_clean = df_clean[df_clean['Amount'].notna()]
    cleaning_report['rows_removed']['null_or_invalid_amount'] = null_amount
    
    
    # ==================== REMOVE DUPLICATES ====================
    # Check for exact duplicates
    duplicates = df_clean.duplicated().sum()
    df_clean = df_clean.drop_duplicates()
    cleaning_report['rows_removed']['exact_duplicates'] = duplicates
    
    
    # ==================== SORT DATA ====================
    df_clean = df_clean.sort_values(['Country', 'Indicator', 'Time']).reset_index(drop=True)
    
    
    # ==================== FINAL REPORT ====================
    cleaning_report['final_rows'] = len(df_clean)
    cleaning_report['total_rows_removed'] = initial_rows - len(df_clean)
    cleaning_report['data_retention_rate'] = f"{(len(df_clean) / initial_rows * 100):.2f}%"
    
    return df_clean, cleaning_report


def print_cleaning_report(report):
    """
    Print a formatted cleaning report.
    
    Parameters:
    -----------
    report : dict
        Cleaning report from clean_economic_data
    """
    print("=" * 60)
    print("DATA CLEANING REPORT")
    print("=" * 60)
    print(f"\nInitial rows: {report['initial_rows']:,}")
    print(f"Final rows: {report['final_rows']:,}")
    print(f"Total rows removed: {report['total_rows_removed']:,}")
    print(f"Data retention rate: {report['data_retention_rate']}")
    
    print("\n" + "-" * 60)
    print("ROWS REMOVED BY REASON:")
    print("-" * 60)
    for reason, count in report['rows_removed'].items():
        if count > 0:
            print(f"  {reason.replace('_', ' ').title()}: {count:,}")
    
    print("\n" + "-" * 60)
    print("STANDARDIZATIONS APPLIED:")
    print("-" * 60)
    for category, mapping in report['standardizations'].items():
        if category == 'currency_fills_by_unit':
            print(f"\n  Currency Fills by Unit:")
            total_fills = 0
            for unit, count in mapping.items():
                print(f"    {unit}: {count:,}")
                total_fills += count
            print(f"    TOTAL: {total_fills:,}")
        elif category == 'monetary_values_inferred':
            print(f"  Monetary currencies inferred from country: {mapping:,}")
        elif isinstance(mapping, dict) and mapping and category != 'currency_fills_by_unit':
            print(f"\n  {category.replace('_', ' ').title()}:")
            for old, new in mapping.items():
                print(f"    '{old}' → '{new}'")
        elif isinstance(mapping, int) and mapping > 0:
            print(f"  {category.replace('_', ' ').title()}: {mapping}")
    
    # Show warnings if any
    if 'warnings' in report and report['warnings']:
        print("\n" + "-" * 60)
        print("⚠️  WARNINGS:")
        print("-" * 60)
        for warning, value in report['warnings'].items():
            if value > 0:
                print(f"  {warning.replace('_', ' ').title()}: {value:,}")
                if warning == 'unknown_monetary_currencies':
                    print(f"    → These monetary values lack currency information")
                    print(f"    → Consider manual review or additional data sources")
    
    print("\n" + "=" * 60)


def query_economic_data(df, country=None, indicator=None, start_date=None, end_date=None, frequency=None):
    """
    Query economic data with flexible filtering options.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        The cleaned economic dataset
    country : str or list of str, optional
        Country name(s) to filter. If None, returns all countries.
    indicator : str or list of str, optional
        Indicator name(s) to filter. If None, returns all indicators.
    start_date : str or datetime, optional
        Start date for time filtering (format: 'YYYY-MM-DD')
    end_date : str or datetime, optional
        End date for time filtering (format: 'YYYY-MM-DD')
    frequency : str, optional
        Data frequency ('Yearly', 'Monthly', 'Quarterly')
        
    Returns:
    --------
    pandas.DataFrame
        Filtered dataset
    """
    result = df.copy()
    
    # Filter by country
    if country is not None:
        if isinstance(country, str):
            result = result[result['Country'] == country]
        elif isinstance(country, list):
            result = result[result['Country'].isin(country)]
    
    # Filter by indicator
    if indicator is not None:
        if isinstance(indicator, str):
            result = result[result['Indicator'] == indicator]
        elif isinstance(indicator, list):
            result = result[result['Indicator'].isin(indicator)]
    
    # Filter by frequency
    if frequency is not None:
        result = result[result['Frequency'] == frequency]
    
    # Filter by date range
    if start_date is not None:
        result = result[result['Time'] >= pd.to_datetime(start_date)]
    if end_date is not None:
        result = result[result['Time'] <= pd.to_datetime(end_date)]
    
    return result.reset_index(drop=True)


def get_available_indicators(df, country=None):
    """Get list of available indicators, optionally filtered by country."""
    if country is not None:
        df_filtered = df[df['Country'] == country]
    else:
        df_filtered = df
    return sorted(df_filtered['Indicator'].unique().tolist())


def get_available_countries(df, indicator=None):
    """Get list of available countries, optionally filtered by indicator."""
    if indicator is not None:
        df_filtered = df[df['Indicator'] == indicator]
    else:
        df_filtered = df
    return sorted(df_filtered['Country'].unique().tolist())


def get_data_summary(df):
    """
    Get a comprehensive summary of the dataset.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        The economic dataset
        
    Returns:
    --------
    dict
        Summary statistics
    """
    summary = {
        'total_records': len(df),
        'countries': df['Country'].nunique(),
        'indicators': df['Indicator'].nunique(),
        'date_range': (df['Time'].min(), df['Time'].max()),
        'frequencies': df['Frequency'].value_counts().to_dict(),
        'country_list': sorted(df['Country'].unique().tolist()),
        'indicator_list': sorted(df['Indicator'].unique().tolist())
    }
    return summary


df = pd.read_csv('10Alytics Hackathon- Fiscal Data.csv')
# Clean the data
df_clean, report = clean_economic_data(df)
print_cleaning_report(report)
    
    # Get summary
summary = get_data_summary(df_clean)
print(f"\nCountries: {summary['countries']}")
print(f"Indicators: {summary['indicators']}")
print(f"Date range: {summary['date_range'][0]} to {summary['date_range'][1]}")
    
    # Query examples
    # egypt_gdp = query_economic_data(df_clean, country='Egypt', indicator='GDP Growth Rate')
    # africa_inflation = query_economic_data(df_clean, indicator='Inflation Rate', start_date='2020-01-01')
```
